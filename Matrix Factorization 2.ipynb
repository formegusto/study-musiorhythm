```python
import pandas as pd
import numpy as np
from sklearn.decomposition import TruncatedSVD
```

```python
# 사용자-영화평점 기반 데이터
rating_data = pd.read_csv("./moviedata/ratings.csv")

# 영화 정보 데이터
movie_data = pd.read_csv("./moviedata/movies.csv")
```

```python
user_movie_rating = rating_data.pivot(
    index="userId",
    columns="movieId",
    values="rating"
).fillna(0)
```

```python
user_movie_rating
```

```python
matrix = user_movie_rating.to_numpy()
```

```python
user_ratings_mean = np.mean(matrix, axis=1)
```

```python
matrix_user_mean = matrix - user_ratings_mean.reshape(-1,1)
```

```python
matrix
```

```python
pd.DataFrame(matrix_user_mean, columns = user_movie_rating.columns).head()
```

```python
from scipy.sparse.linalg import svds
```

```python
U, sigma, Vt = svds(matrix_user_mean, k = 12)
```

```python
print(U.shape)
print(sigma.shape)
print(Vt.shape)
```

```python
sigma = np.diag(sigma)
```

```python
sigma.shape
```

```python
sigma[0]
```

```python
svd_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.reshape(-1, 1)
```

```python
df_svd_preds = pd.DataFrame(svd_user_predicted_ratings, columns = user_movie_rating.columns)
```

```python
def recommend_movies(df_svd_preds, user_id, ori_movies_df, ori_ratings_df, num_recommentations=5):
        user_row_number = user_id - 1
        sorted_user_predictions = df_svd_preds.iloc[user_row_number].sort_values(ascending=False)
        user_data = ori_ratings_df[ori_ratings_df.userId == user_id]
        
        user_history = user_data.merge(ori_movies_df, on="movieId").sort_values(['rating'], ascending=False)
        
        recommendations = ori_movies_df[~ori_movies_df['movieId'].isin(user_history['movieId'])]
        recommendations = recommendations.merge(pd.DataFrame(sorted_user_predictions).reset_index(), on="movieId")
        recommendations = recommendations.rename(columns={user_row_number: "Predictions"}).sort_values("Predictions", ascending=False)
        
        return user_history, recommendations
```

```python
already_rated, predictions = recommend_movies(df_svd_preds, 330, movie_data, rating_data, 10)
```
