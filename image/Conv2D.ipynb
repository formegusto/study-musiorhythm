```python
x = np.round(np.random.rand(4,28,28) * 10)
conv_x = x.reshape(4, 28, 28, 1)
```

```python
model = tf.keras.models.Sequential()
model.add(
    tf.keras.layers.Conv2D(
        32, (3,3), activation='linear', input_shape=conv_x.shape[1:],
        kernel_initializer=tf.keras.initializers.Constant(value=1)
    )
)
model.add(
    tf.keras.layers.MaxPooling2D((2,2))
)
model.add(
    tf.keras.layers.Conv2D(
        64, (3,3), activation='linear', input_shape=conv_x.shape[1:],
        kernel_initializer=tf.keras.initializers.Constant(value=1)
    )
)
model.add(
    tf.keras.layers.MaxPooling2D((2,2))
)
model.add(
    tf.keras.layers.Conv2D(
        64, (3,3), activation='linear', input_shape=conv_x.shape[1:],
        kernel_initializer=tf.keras.initializers.Constant(value=1)
    )
)
model.add(tf.keras.layers.Flatten())
model.add(
    tf.keras.layers.Dense(64, activation='linear', input_shape=conv_x.shape[1:],
        kernel_initializer=tf.keras.initializers.Constant(value=1))
)

class MyDenseLayer(tf.keras.layers.Layer):
  def __init__(self, num_outputs):
    super(MyDenseLayer, self).__init__()
    self.num_outputs = num_outputs

  def build(self, input_shape):
    self.kernel = self.add_weight("kernel",
                                  shape=[int(input_shape[-1]),
                                         self.num_outputs])

  def call(self, inputs):
    return tf.round(inputs / 1000000) 

model.add(MyDenseLayer(64))
model.add(tf.keras.layers.Dense(10, activation="softmax"))

model(conv_x)
```

```python
y = tf.keras.layers.Conv2D(
    2, (3,3), activation='linear', input_shape=conv_x.shape[1:],
    kernel_initializer='one'
)
print(y(conv_x).numpy())
```
